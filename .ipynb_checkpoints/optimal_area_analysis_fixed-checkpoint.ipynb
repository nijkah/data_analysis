# Optimal Area Analysis about medical tour

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
path = 'C:\workspace'
os.chdir(path)
```

```python
for file in sorted(os.listdir()):
    print(file)
```

```python
hospitals = pd.read_csv('dropped_hospitals.csv', encoding='utf-8')
print(hospitals.columns)
hospitals =hospitals[['VIATR_FCLTY_CD','VIATR_FCLTY_NM','ADRES_NM','lat','lon']]
```

```python
hospitals.head(5)
```

```python
len(hospitals)
```

```python
adr = hospitals[['lat','lon']]
```

```python
adr.plot(x=1, y=0, kind='scatter', marker='o', title='hospitals')
```

```python
np_mat = adr.as_matrix()
```

```python
model = KMeans(n_clusters=10, init='random').fit(np_mat)
centroids= model.cluster_centers_
```

```python
print(centroids)
```

```python
centroids = [[  37.49497644,  126.9087597 ],\
 [  37.48921727,  126.96988735],\
 [  37.52094002,  127.11896207],\
 [  37.58056525,  127.0763558 ],\
 [  37.58874354,  126.92000749],\
 [  37.53350116,  126.85076075],\
 [  37.64705373,  127.04938081],\
 [  37.5811437,   127.02202841],\
 [  37.50653759,  127.03483601],\
 [  37.55729352,  126.96571228]]
```

```python
print(centroids)
```

```python
import matplotlib.cm as cm
x = np.arange(10)
ys = [i+x+(i*x)**2 for i in range(10)]

colors = cm.rainbow(np.linspace(0,1,len(ys)))
```

```python
for i in range(10):
    plt.scatter(i+1,1, marker='o', c=colors[i] )
plt.title('Color Scheme')
plt.xticks(list(i+1 for i in range(10)))
plt.show()
```

# K - Means Clustering

```python
for i in range(10):
    plt.scatter(np_mat[model.labels_==i,1],np_mat[model.labels_==i,0], marker='.', c=colors[i] )
axes =plt.gca()
xlim = axes.get_xlim()
ylim = axes.get_ylim()
plt.show()
print(xlim, ylim)
```

```python
rank = list()
for i in range(10):
    print('{} area: {} hospitals'.format(i+1,np_mat[model.labels_==i].shape[0]))
    rank.append((i+1,np_mat[model.labels_==i].shape[0]/np_mat.shape[0]))
```

```python
rank
```

```python
rank_num = [i[0] for i in rank]
rank_val = [i[1] for i in rank]
```

```python
np_mat
```

```python
model.labels_
```

## Give scores by the number of hospitals

```python
adr['class'] = model.labels_
scores = pd.DataFrame([i+1 for i in range(10)],columns=['class'])
```

```python
scores['score'] = scores['class'].replace(rank_num, rank_val)
scores
```

```python
for i in range(10):
    plt.bar(i+1, scores['score'][i], color=colors[i])
plt.xticks(list(i+1 for i in range(10)))
plt.show()
```

```python
for i in range(10):
    plt.scatter(centroids[i][1], centroids[i][0], c=colors[i])
plt.xlim(xlim)
plt.ylim(ylim)
plt.show()
```

## Show centroid of each area

```python
for i in range(10):
    plt.scatter(np_mat[model.labels_==i,1],np_mat[model.labels_==i,0], marker='.', c='gray' )    #1
for i in range(10):
    plt.scatter(centroids[i][1], centroids[i][0], marker='*', s=70, c=colors[i])
plt.xlim(xlim)
plt.ylim(ylim)
plt.show()
```

```python
dev_area =pd.read_csv('dev_area.csv', encoding='utf-8')
dev_area = dev_area.drop('Unnamed: 0', axis=1)
print(len(dev_area))
dev_area.head(5)
```

# Core Commercial Area

```python
dev_area.plot(x=2,y=1,  xlim=xlim, ylim=ylim, kind='scatter', marker='o', title='developed_areas')
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=1)
```

```python
dev_area_mat = dev_area[['lat','lon']].as_matrix()
classifier.fit(centroids, [i+1 for i in range(10)])
prediction = classifier.predict(dev_area_mat)
```

```python
for i in range(10):
    plt.scatter(dev_area_mat[prediction==i+1,1],dev_area_mat[prediction==i+1,0], marker='.', c=colors[i] )
plt.xlim(xlim)
plt.ylim(ylim)
plt.show()
```

```python
for i in range(10):
    print(len(dev_area_mat[prediction==i])/len(dev_area_mat))
    scores['score'][i] = scores['score'][i] + len(dev_area_mat[prediction==i+1])/len(dev_area_mat)
```

```python
scores
```

```python
for i in range(10):
    plt.bar(i+1, scores['score'][i], color=colors[i])
plt.xticks(list(i+1 for i in range(10)))
plt.show()
```
